```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
I = 2.0
K = 400.0
A = 0.0175
TIMER_DELT = 0.005
FINTIM = 1.5
PRDEL = 0.05
D_values = [5.656, 16.968, 39.592, 56.56, 113.12]

# Initialize variables
HEAD = 0.0
ANGVEL = 0.0
ANGACC = 0.0
ATIME = 0.0

# Lists to store values for plotting
head_values = []

# Simulation loop
for t in np.arange(0, FINTIM, TIMER_DELT):
    # Calculate the desired heading (INPUT) as a function of time
    INPUT = A * t

    # Calculate the error
    ERROR = INPUT - HEAD

    # Calculate the control torque
    TORQUE = K * ERROR - D_values[0] * ANGVEL  # Change D_values index for different runs

    # Calculate angular acceleration
    ANGACC = TORQUE / I

    # Update HEAD and ANGVEL using numerical integration
    HEAD += ANGVEL * TIMER_DELT
    ANGVEL += ANGACC * TIMER_DELT

    # Append the current HEAD value to the list
    head_values.append(HEAD)

# Plotting
plt.plot(np.arange(0, FINTIM, TIMER_DELT), head_values)
plt.xlabel('Time')
plt.ylabel('HEAD')
plt.title('HEADING VS TIME')
plt.grid()
plt.show()
```
